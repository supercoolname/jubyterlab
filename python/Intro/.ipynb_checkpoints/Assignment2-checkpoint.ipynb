## Part 1
The following code loads the olympics dataset (olympics.csv), which was derrived from the Wikipedia entry on [All Time Olympic Games Medals](https://en.wikipedia.org/wiki/All-time_Olympic_Games_medal_table), and does some basic data cleaning. 

The columns are organized as # of Summer games, Summer medals, # of Winter games, Winter medals, total # number of games, total # of medals. Use this dataset to answer the questions below.

```python
import pandas as pd

df = pd.read_csv('olympics.csv', index_col=0, skiprows=1)

for col in df.columns:
    if col[:2]=='01':
        df.rename(columns={col:'Gold'+col[4:]}, inplace=True)
    if col[:2]=='02':
        df.rename(columns={col:'Silver'+col[4:]}, inplace=True)
    if col[:2]=='03':
        df.rename(columns={col:'Bronze'+col[4:]}, inplace=True)
    if col[:1]=='№':
        df.rename(columns={col:'#'+col[1:]}, inplace=True)

print(df.head())

names_ids = df.index.str.split('\xc2\xa0\(') # split the index by '('

df.index = names_ids.str[0] # the [0] element is the country name (new index) 
df['ID'] = names_ids.str[1].str[:3] # the [1] element is the abbreviation or ID (take first 3 characters from that)

df.index = names_ids.str[0]

df = df.drop('Totals')

df.head()
```

### Question 4
Write a function that creates a Series called "Points" which is a weighted value where each gold medal (`Gold.2`) counts for 3 points, silver medals (`Silver.2`) for 2 points, and bronze medals (`Bronze.2`) for 1 point. The function should return only the column (a Series object) which you created, with the country names as indices.

```python
census_df = pd.read_csv('census.csv')
census_df.head()
```

```python
def answer_five(census_df):
    st_count=pd.Series();
    census_df = census_df.set_index(['STNAME'])
    print(census_df.head())

    for st in census_df.index.unique():
        ct_count = len(census_df.loc[st,'CTYNAME'].unique())
        st_count = st_count.append(pd.Series({'STNAME': st,
                        'CITY_COUNT': ct_count}))
    return st_count['CITY_COUNT'].max()

print(census_df.index)
print ("result is")
print(answer_five(census_df))
print ("result done")
print(census_df.index)
census_df.head()
```

### Question 7
Which county has had the largest absolute change in population within the period 2010-2015? (Hint: population values are stored in columns POPESTIMATE2010 through POPESTIMATE2015, you need to consider all six columns.)

e.g. If County Population in the 5 year period is 100, 120, 80, 105, 100, 130, then its largest change in the period would be |130-80| = 50.

*This function should return a single string value.*

```python
def find_max_diff(row):
    #print(row)
    data = row[['STNAME',
                   'CTYNAME',
                   'BIRTHS2010','BIRTHS2011','BIRTHS2012','BIRTHS2013','BIRTHS2014','BIRTHS2015']]
    max_diff = 0
    for i in range(2,8):
        for j in range(3,8):
            if(abs(data.iloc[i]-data.iloc[j]) > max_diff):
                max_diff = abs(data.iloc[i]-data.iloc[j])
    row['MAX_DIFF'] = max_diff
    return row

def answer_seven(census_df):
    columns_to_keep = ['STNAME',
                   'CTYNAME',
                   'BIRTHS2010','BIRTHS2011','BIRTHS2012','BIRTHS2013','BIRTHS2014','BIRTHS2015']
    census_df =census_df[columns_to_keep].where(census_df['SUMLEV'] == 50)
                
    census_df = census_df.apply(find_max_diff,axis=1)
    print(census_df.head())
    
    census_df = census_df.sort_values(['MAX_DIFF'],ascending=False)
    print(census_df.head())
    return census_df.iloc[0]['CTYNAME']

print ("result is")
print(answer_seven(census_df))
print ("result done")
census_df.head()
```

### Question 8
In this datafile, the United States is broken up into four regions using the "REGION" column. 

Create a query that finds the counties that belong to regions 1 or 2, whose name starts with 'Washington', and whose POPESTIMATE2015 was greater than their POPESTIMATE 2014.

*This function should return a 5x2 DataFrame with the columns = ['STNAME', 'CTYNAME'] and the same index ID as the census_df (sorted ascending by index).*

```python
def start_with_name(row):
    data = row[['INDEX_ID','STNAME','CTYNAME']]
    washington = data[2].startswith('Washington')
    if(washington):
        data['Washington'] = True
    else:
        data['Washington'] = False
    return data

def answer_eight(census_df):
    census_df['INDEX_ID'] = census_df.index
    columns_to_keep = [ 'SUMLEV','REGION','STNAME',
                   'CTYNAME',
                   'POPESTIMATE2014','POPESTIMATE2015', 'INDEX_ID']
    census_df =census_df[columns_to_keep].where((census_df['SUMLEV'] == 50) & ((census_df['REGION'] == 1) | (census_df['REGION'] == 2)))
    census_df =census_df.where(census_df['POPESTIMATE2015'] > census_df['POPESTIMATE2014']).dropna()
    mask = census_df.apply(start_with_name,axis=1)
    mask = mask.where(mask['Washington']== True).dropna()
    print("mask: ",mask)
    
    result = pd.merge(mask,census_df, how= 'left', left_on=['INDEX_ID'], right_on = ['INDEX_ID'])
    print("result: ", result)
    result = result.set_index('INDEX_ID')
    print("result set_index: ", result)
    result = result[['STNAME_x','CTYNAME_x']]
    print("result result: ", result)
    return result

print ("result is")
print(answer_eight(census_df))
print ("result done")
```
